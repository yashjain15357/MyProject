```python
# C:\Users\YASH JAIN\Documents\New Folder\OneDrive\Desktop\CodeBox\python language\panda\data\purchase_data_exe.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\YASH JAIN\Documents\New Folder\OneDrive\Desktop\data\purchase_data_exe.csv")
df.head()
```

```python
df["date"]=pd.to_datetime(df["date"], format="%d/%m/%Y")
df["day"] = df["date"].dt.day_name()
df.head()
```

```python
df["payment_method"].unique()
```

```python
x=df["payment_method"].value_counts().index
y= df["payment_method"].value_counts().values
print(y)
plt.figure(figsize=(8,4))
plt.bar(x,y,color=["c","m"],width=0.3,label=x)
plt.legend()
plt.xticks(rotation=45,)
plt.title("Payment Method",fontsize=18,fontweight="bold")
plt.xlabel("Payment Types",fontsize=14)
plt.ylabel("Count of Payment type",fontsize=14)

for i, v in enumerate(y):
    plt.text(i, v + 1, str(v), ha='center', fontsize=10, fontweight="bold", color="black")
plt.show()
```

```python
print(df["date"].value_counts(sort=False))
dta=df["date"].value_counts(sort=False)
x=dta.index

y=dta.values
plt.bar(x,y)
plt.title("Customer" ,fontsize=18,fontweight='bold')
plt.xlabel("Date",fontsize=14)
plt.ylabel("Countumer count",fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
print(df["day"].value_counts(sort=False))
dta=df["day"].value_counts(sort=False)
x=dta.index

y=dta.values
plt.bar(x,y)
plt.title("Customer" ,fontsize=18,fontweight='bold')
plt.xlabel("Date",fontsize=14)
plt.ylabel("Countumer count",fontsize=14)
plt.xticks(rotation=45)


# i → The x-coordinate (position of the bar).
# v + 1 → The y-coordinate where the text will be placed (slightly above the bar).
# str(v) → The actual text (count value) to display.
for i, v in enumerate(y):
    plt.text(i, v + 1, str(v), ha='center', fontsize=10, fontweight="bold", color="black")
plt.show()
```

```python
dat = df["value [USD]"]
# Define x-tick positions and labels
tick_positions = [i for i in range(0,2500,100)]  # Numeric tick positions
tick_labels = [f"{i}-{i+100}" for i in tick_positions]  # String labels for each range

# Create plot
plt.figure(figsize=(16, 8))
sns.histplot(dat, bins=25, color="r", alpha=0.6,kde=True)

# Set x-axis ticks correctly
plt.xticks(ticks=tick_positions, labels=tick_labels, rotation=45)
plt.title("")
plt.xlabel("Price Bins (USD)")
plt.ylabel("Frequency of Transactions")

plt.show()
```

```python
x=df["product_category"].value_counts(sort=False).values
y=df["product_category"].value_counts(sort=False).index
explode = list([0.1] * len(y)) 
explode[0]=0.3
plt.pie(x,radius=1.5,
        labels=y,
        colors = ['gold', 'lightblue', 'lightcoral',"#94d325","#2564d3","#b125d3"],
        autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'},
        explode =explode)
plt.title("Product Sell", loc='left', x=1, y=1,fontsize=18,fontweight="bold")

plt.show()
```
